# Write a Python code snippet to generate a histogram of 10,000 samples from a normal distribution with mean 0 and standard deviation 1. Explain why the shape of the histogram resembles a bell curve.

```python
import pandas as pd
import numpy as np
```

```python
#Generate a histogram of 10,000 samples from a normal distribution with mean 0 and standard deviation 1.

samples = np.random.normal(loc=0, scale = 1, size = 10000)
```

```python
# Plotting Histogram
df = pd.DataFrame(samples, columns = ['Samples'])

df['Samples'].plot.hist(bins = 50, color = 'r', title = 'Histogram of 10000 samples from a normal distribution', xlabel = 'Values')
```

```python
print('Explain why the shape of the histogram resembles a bell curve.:\nThe shape of the histogram resembles a bell curve because it represents a normal distribution, which is symmetric about the mean (0) and tapers off as values move away from the mean. This is characteristic of the normal distribution, often called a "bell curve" due to its shape.')
```

# Use Python to load the dataset (score.csv) and plot the distribution of the 'Income' column. Describe its shape and explain why it is not normally distributed.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('score.csv')
```

```python
df['Income'].plot.hist(bins=50, alpha=0.6, color='b', title='Distribution of Income', xlabel = 'Income', ylabel = 'Frequency')
```

```python
print('The Shape of this graph is Right-Skewed \n')

print('****************************************\n')

print('Explaination:\nIncome distributions are typically not normally distributed because wealth is not evenly distributed across populations. Most individuals cluster around lower or median incomes, while a smaller proportion has much higher incomes, leading to a right skew in the data.')
```
