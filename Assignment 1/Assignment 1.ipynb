```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("E:\Programming\Mastering Statistics For Data Science\Employee_Salaries_Data.csv")
data.head(30)
```

```python
Q1 = data['Years of Experience'].quantile(0.25)
Q3 = data['Years of Experience'].quantile(0.75)
Experience_IQR = Q3 - Q1
print(f"the Interquartile Range (IQR) for the years of experience is: {Experience_IQR}")
```

```python
Experience_Range = data['Years of Experience'].max() - data['Years of Experience'].min()
print(Experience_Range)
```

```python
comparison_ratio = Experience_IQR/Experience_Range

print(f'The Comparison of IQR to the range of year of Experience is :{comparison_ratio}')
```

# The employee tenure distribution is moderately concentrated around the middle 50%, with some variability but not extreme. The relatively small IQR compared to the overall range points to the likelihood of a few outliers (either very experienced or very new employees), but for the most part, employees have a similar level of experience.

```python
# Identify the employee with the maximum salary deviation from the median salary.

median_salary = data['Salary'].median()
# print(f'The median salary is : {median_salary}')

# Absolute Salary Deviation:
data["Salary_Deviation"] = abs(data['Salary'] - median_salary)

# Maximum Salary Deviation:
maxSalary = data["Salary_Deviation"].max()

maxSalary_deviation = data.loc[data['Salary_Deviation'].idxmax(), 'Employee ID']


print(f"The Employee with Maximum Salary Deviation is : {maxSalary_deviation}, with the salary of : {maxSalary}")
```

```python
# the range of the salaries across all employees.

salary_range = data['Salary'].max() - data['Salary'].min()
print(f'The range of the salaries across all employees : {salary_range}')
```

```python
# Compare this deviation to the range of the salaries across all employees.

salaryComparison_ratio = (maxSalary) / (salary_range)

print(f"The Salary Comparison Ratio is : {salaryComparison_ratio}")
```

It shows how far the employee with the maximum deviation is from the median salary compared to the total range of salaries.
